```python
# Dependecy
import pandas as pd
```

```python
# Read the billboard data into pandas
hot_100 = pd.read_csv('data/hot_100_1958_2019.csv')

# Display dataframe 
hot_100
```

```python
# Show column data types
hot_100.dtypes
```

```python
# Convert 'WeekID' from a string to a datetime object
hot_100['WeekID'] = pd.to_datetime(hot_100['WeekID'])
```

```python
# Show column data types
hot_100.dtypes
```

```python
# Sort dataframe by 'WeekID'
sorted_hot_100 = hot_100.sort_values(['WeekID'], ascending=[False])
sorted_hot_100
```

```python
# Delete 'url' column
del sorted_hot_100['url']
```

```python
# Re-index the dataframe
clean_hot_100 = sorted_hot_100.reset_index(drop=True)

# Display dataframe
clean_hot_100
```

```python
# Rename columns
cleaned_billboard_data = clean_hot_100.rename(columns={'WeekID': 'week_id', 'Week Position': 'week_pos', 'Song': 'track', 'Performer': 'artist', 'SongID': 'track_id', 'Instance': 'instance', 'Previous Week Position': 'prev_week_pos', 'Peak Position': 'peak_pos', 'Weeks on Chart': 'weeks_on_chart'})
cleaned_billboard_data.head()
```

```python
cleaned_billboard_data = cleaned_billboard_data
```

```python
# Read and store content of the excel file  
hot_100_audio = pd.read_excel ('data/hot_100_audio_features.xlsx') 
```

```python
# Write the dataframe object into csv file 
hot_100_audio.to_csv ('data/hot_100_audio_features.csv', index = None, header=True) 
```

```python
# Read csv file and convert into a dataframe object 
audio_features = pd.DataFrame(pd.read_csv('data/hot_100_audio_features.csv')) 
  
# Display dataframe 
audio_features
```

```python
# Loop to print all column names for easier reading
for col in audio_features.columns: 
    print(col)
```

```python
# Delete 'spotify_track_preview_url' and 'spotify_track_album' columns
del audio_features['spotify_track_preview_url']
```

```python
audio_features.head()
```

```python
cleaned_audio_features_data = audio_features.rename(columns={'Song': 'track', 'Performer': 'artist', 'SongID': 'track_id'})
cleaned_audio_features_data.head()
```

```python
cleaned_audio_features_data = cleaned_audio_features_data
```

```python
# Export cleaned dataframes as csvs
cleaned_billboard_data.to_csv('data/billboard_hot_100_1958_2019.csv', index = False)
cleaned_audio_features_data.to_csv('data/audio_features_hot_100_1958_2019.csv', index = False)
```
