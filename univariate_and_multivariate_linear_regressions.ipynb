```python
import numpy as np
import matplotlib.pyplot as plt  
import pandas as pd
```

```python
data = pd.read_csv('data/grouped_audio_billboard_genre_data.csv')
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
cols_to_norm = ['spotify_track_explicit','spotify_track_duration_ms','spotify_track_popularity','danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','time_signature','week_pos','instance','peak_pos']
data[cols_to_norm] = StandardScaler().fit_transform(data[cols_to_norm])
data
```

```python
# Assign X (data) and y (target)
X = data.peak_pos.values.reshape(-1, 1)
y = data.weeks_on_chart.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
# Create a LinearRegression model and fit it to the scaled training data
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)
```

```python
predictions = model.predict(X)
print(f"True output: {y[0]}")
print(f"Predicted output: {predictions[0]}")
print(f"Prediction Error: {predictions[0]-y[0]}")
```

```python
x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
plt.scatter(X, y, c='lightblue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
```

```python
X1 = data[['spotify_track_popularity','danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','time_signature','week_pos','instance','peak_pos']]
y1 = data.weeks_on_chart.values.reshape(-1, 1)

print("Shape: ", X1.shape, y1.shape)
```

```python
# Split the data into training and testing

from sklearn.model_selection import train_test_split

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, random_state=98)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data

X1_scaler = StandardScaler().fit(X1_train)
y1_scaler = StandardScaler().fit(y1_train)
```

```python
 # Transform the training and testing data using the X_scaler and y_scaler models

X1_train_scaled = X1_scaler.transform(X1_train)
X1_test_scaled = X1_scaler.transform(X1_test)
y1_train_scaled = y1_scaler.transform(y1_train)
y1_test_scaled = y1_scaler.transform(y1_test)
```

```python
# Create a LinearRegression model and fit it to the scaled training data

model = LinearRegression()
model.fit(X1_train_scaled, y1_train_scaled)
training_score = model.score(X1_train_scaled, y1_train_scaled)
testing_score = model.score(X1_test_scaled, y1_test_scaled)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# Make predictions using a fitted model
# Plot the difference between the model predicted values and actual y values, versus the model predicted values

predictions = model.predict(X1_test_scaled)
model.fit(X1_train_scaled, y1_train_scaled)
plt.scatter(model.predict(X1_train_scaled), model.predict(X1_train_scaled) - y1_train_scaled, c="blue", label="Training Data")
plt.scatter(model.predict(X1_test_scaled), model.predict(X1_test_scaled) - y1_test_scaled, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y1_test_scaled.min(), xmax=y1_test_scaled.max())
plt.title("Residual Plot")
plt.show()
```
